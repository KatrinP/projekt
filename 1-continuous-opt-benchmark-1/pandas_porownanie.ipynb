```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# import matplotlib
# matplotlib.style.use('ggplot')
%matplotlib inline
import os, glob
```

```python
filenames = glob.glob('*.txt')
```

```python
dfs = []
for filename in filenames:
    # use this if column names are in the csv file already
    # (not the case for the data in the example dataset
    # data = pandas.io.parsers.read_csv(filename)  
    # use this if column names are missing as in the example dataset
    names = ['Iteration', 'fitness', 'vectors']  
    data = pd.io.parsers.read_csv(filename, header=None, names=names, delimiter=';')
    dfs.append(data)
# Concatenate all data into one DataFrame
df = pd.concat(dfs, ignore_index=True)
```
